```python
import pandas as pd
import matplotlib.pyplot as plt

# Importar os dados do CSV
df = pd.read_csv('./popu.csv')

# Filtrar as colunas necessárias
df = df[['Country Name', '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]']]

# Converter as colunas de taxa de crescimento populacional para tipo numérico
for year in ['2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]']:
    df[year] = pd.to_numeric(df[year], errors='coerce')

"""
Análise de Crescimento Populacional

Estamos analisando a taxa de crescimento populacional anual em cinco países (Brasil, China, França, Estados Unidos e Emirados Árabes Unidos) nos anos de 2012 a 2016.

Comparar a taxa de crescimento populacional anual de 2012 a 2016.

Os gráficos mostram a evolução da taxa de crescimento populacional ao longo dos anos de 2012 a 2016.
"""

# Configurar os dados para visualização
countries = df['Country Name']
growth_2012 = df['2012 [YR2012]']
growth_2013 = df['2013 [YR2013]']
growth_2014 = df['2014 [YR2014]']
growth_2015 = df['2015 [YR2015]']
growth_2016 = df['2016 [YR2016]']

# Gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(countries, growth_2012, marker='o', label='2012')
plt.plot(countries, growth_2013, marker='o', label='2013')
plt.plot(countries, growth_2014, marker='o', label='2014')
plt.plot(countries, growth_2015, marker='o', label='2015')
plt.plot(countries, growth_2016, marker='o', label='2016')
plt.title('Taxa de Crescimento Populacional Anual (%) de 2012 a 2016')
plt.xlabel('País')
plt.ylabel('Taxa de Crescimento Populacional (%)')
plt.legend()
plt.grid(True)
plt.show()

# Gráfico de barra
plt.figure(figsize=(10, 6))
bar_width = 0.15
index = range(len(countries))
plt.bar(index, growth_2012, bar_width, color='skyblue', label='2012')
plt.bar([i + bar_width for i in index], growth_2013, bar_width, color='orange', label='2013')
plt.bar([i + bar_width * 2 for i in index], growth_2014, bar_width, color='green', label='2014')
plt.bar([i + bar_width * 3 for i in index], growth_2015, bar_width, color='red', label='2015')
plt.bar([i + bar_width * 4 for i in index], growth_2016, bar_width, color='purple', label='2016')
plt.xlabel('País')
plt.ylabel('Taxa de Crescimento Populacional (%)')
plt.title('Taxa de Crescimento Populacional Anual (%) de 2012 a 2016')
plt.xticks([i + bar_width * 2 for i in index], countries)
plt.legend()
plt.show()

# Histograma
plt.figure(figsize=(10, 6))
plt.hist([growth_2012, growth_2013, growth_2014, growth_2015, growth_2016], bins=5, color=['green', 'orange', 'red', 'blue', 'purple'], edgecolor='black', label=['2012', '2013', '2014', '2015', '2016'])
plt.title('Distribuição da Taxa de Crescimento Populacional de 2012 a 2016')
plt.xlabel('Taxa de Crescimento Populacional (%)')
plt.ylabel('Frequência')
plt.legend()
plt.show()

# Gráfico de pizza
plt.figure(figsize=(10, 6))
plt.pie(growth_2016, labels=countries, autopct='%1.1f%%', startangle=140)
plt.title('Proporção da Taxa de Crescimento Populacional por País em 2016')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Gráfico scatter
plt.figure(figsize=(10, 6))
plt.scatter(countries, growth_2012, color='blue', label='2012')
plt.scatter(countries, growth_2013, color='orange', label='2013')
plt.scatter(countries, growth_2014, color='green', label='2014')
plt.scatter(countries, growth_2015, color='red', label='2015')
plt.scatter(countries, growth_2016, color='purple', label='2016')
plt.title('Taxa de Crescimento Populacional Anual (%) de 2012 a 2016')
plt.xlabel('País')
plt.ylabel('Taxa de Crescimento Populacional (%)')
plt.legend()
plt.grid(True)
plt.show()
```
